### Scatter and corr

```python
ctcz_jjas_arr = ctcz.precip.values
nino_jjas_arr = jjas_nino.values
```

```python
ctcz.precip.values
```

```python
ctcz_jjas_arr
```

```python
nino_jjas_arr
```

```python
ctcz_anom = ctcz_jjas_arr - ctcz_jjas_arr.mean()
nino_anom = nino_jjas_arr - nino_jjas_arr.mean()
```

```python
ctcz_sdd = ctcz_anom/ctcz_jjas_arr.std()
nino_sdd = nino_anom/nino_jjas_arr.std()
```
